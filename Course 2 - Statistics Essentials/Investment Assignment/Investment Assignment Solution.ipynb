```python
# Importing all the necessary libraries to work with
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
# Loading and Reading companies and rounds2 data into pandas
companies = pd.read_csv("companies.csv", encoding='unicode_escape')
rounds2 = pd.read_csv("rounds2.csv", encoding='unicode_escape')
```

```python
# Checking length and all the datatypes and columns of companies data
companies.info()
```

```python
# Checking unique types of permalink values and duplicacies 
companies.permalink.value_counts()
```

```python
# Checking length and all the dtypes of all the columns for rounds2 data
rounds2.info()
```

```python
# Checking unique types of company_permalink values and duplicacies 
rounds2.company_permalink.value_counts()
```

```python
# Checking data whose company name is solarflare
# These are not the same company but companies having same name or may be subsidaries 
obj = rounds2[rounds2["company_permalink"] == "/ORGANIZATION/SOLARFLARE"]
obj
```

```python
# Checking the company data's first few rows 
companies.head()
```

```python
# Checking the rounds2 data's first few rows 
rounds2.head()
```

```python
# Changing all the companies permalink column to lower case
companies["permalink"] = (companies.permalink.apply(lambda x: x.lower()))
```

```python
# Changing all the rounds2 company_permalink column to lower case and renaming it as permalink
rounds2["permalink"] = (rounds2.company_permalink.apply(lambda x: x.lower()))
```

```python
# Droping column company_permalink 
rounds2.drop(["company_permalink"], axis=1, inplace=True)
```

```python
# Cheching if both the data sets column named permalink has similar values
companies["permalink"].isin(rounds2["permalink"]).value_counts()
```

```python
# Cheching if both the data sets column named permalink has similar values the other way
rounds2["permalink"].isin(companies["permalink"]).value_counts()
```

```python
# Merging both the data sets into single one data frame called master_frame
master_frame = pd.merge(companies, rounds2, on=["permalink"], how="inner")
master_frame.head()
```

```python
# Checking the shape of master_frame
master_frame.shape
```

```python
# Checking unique values of column funging_round_type 
master_frame["funding_round_type"].value_counts()
```

```python
# Grouping the data set by unique values in funding_round_type column
df_by_fund = master_frame.groupby("funding_round_type")
```

```python
# Checking the maximum value by raised_amount_usd column
df_by_fund["raised_amount_usd"].sum()/1000000000
```

```python
# Checking the unique values by funding_round_code column
df_by_fund["funding_round_code"].value_counts()
```

```python
# Grouping by column "country_code"
df_by_cntry = master_frame.groupby("country_code")
```

```python
# Fetching the maximum no. of investment done by a country in decending order
df_by_cntry["name"].count().sort_values(ascending=False)
```

```python
# Grouping by funding_round_type and then country_code
top9 = master_frame.groupby(["funding_round_type", "country_code"])
```

```python
# Fetching the total sum of first ten rows by raised_amount_usd  in decending order 
top9["raised_amount_usd"].sum().sort_values(ascending = False)[:10]
```

```python
# Taking a look at master_frame data set column dtype
master_frame.info()
```

```python
# Taking the first category as the primary category from the category_list column 
master_frame["category_list"] = master_frame["category_list"].apply(lambda x: str(x).split("|")[0])
```

```python
# Cheching effect of the previous operation
master_frame.head()
```

```python
# Loading and Reading the mapping data set
mapping = pd.read_csv("mapping.csv")
```

```python
# Finally merging the data into one data set and keeping only rquired columns for further analysis
df_final = master_frame.merge(mapping, how="inner", on=["category_list"])
```
